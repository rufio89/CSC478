```python
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result
```

```python
vs_target_train = pd.read_csv("trainClasses.txt", sep="\t", header = None)
vs_target_test = pd.read_csv("testClasses.txt", sep="\t", header = None)
vs_target_train = np.array(vs_target_train[1])
vs_target_test = np.array(vs_target_test[1])
vs_target_train
```

```python
def knn_search(x, D, K, measure):
    """ FIND K NEAREST NEIGBORS OF DATA MONG K"""
    if measure == 0:
        #EUCLIDEAN DISTANCES FROM OTHER POINTS
        dists = np.sqrt(((D-x)**2).sum(axis=1))
    elif measure ==1:
        D_norm = np.array([np.linalg.norm(D[i]) for i in range(len(D))])
        x_norm = np.linalg.norm(x)
        sims = np.dot(D,x)/ (D_norm * x_norm)
        dists = 1 - sims
    idx = np.argsort(dists)
    return idx[:K]
```

```python
def classify(vs_new_test, vs_new_train, K, m):
    result_arrays = [knn_search(vs_new_test[i], vs_new_train, K, m) for i in range(len(vs_new_test))]
    results = [vs_target_train[result_arrays[i]] for i in range(len(result_arrays))]
    count_results = []
    predicted_classes =[]
    i = 0
    while i < len(results):
        unique, counts = np.unique(results[i], return_counts=True)
        count_results.insert(i,dict(zip(unique, counts)))
        i=i+1

    i=0
    while i < len(count_results):
        one_count = 0
        zero_count = 0
        if 0 in count_results[i]:
            zero_count = count_results[i][0]
        if 1 in count_results[i]:
            one_count = count_results[i][1]

        if one_count > zero_count:
            predicted_classes.insert(i, 1)
        else:
            predicted_classes.insert(i, 0)
        i=i+1
    predictions = np.array(predicted_classes)
    return predictions

def get_accuracy_ratio(vs_new_test, vs_new_train, K, measure):
    predictions = classify(vs_new_test, vs_new_train, K, measure)
    accuracy = [vs_target_test[i]== predictions[i] for i in range(len(vs_target_test))]
    return float(accuracy.count(True))/len(accuracy)

def test_k(vs_new_test, vs_new_train,K):
    predictions_list = []
    K = K + 1
    for i in range(1,K):
        predictions_list.insert(i,['Euclidean',i,get_accuracy_ratio(vs_new_test, vs_new_train,i,0)])
        predictions_list.insert(i,['Cosine',i,get_accuracy_ratio(vs_new_test, vs_new_train, i,1)])
        
    
    headers =['Measure','K', 'Accuracy']
    return pd.DataFrame(predictions_list, columns=headers).sort_values(by=['Measure', 'K'], ascending=[False,True])

test_k(vs_test_norm, vs_train_norm, 20)
```
