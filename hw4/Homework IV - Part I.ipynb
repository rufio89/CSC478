```python
import numpy as np
import pylab as pl
import pandas as pd
from sklearn import preprocessing
from sklearn.cluster import KMeans 
from sklearn.metrics import completeness_score, homogeneity_score
```

## A. Load Data and Perform Min-Max

```python
data = pd.read_csv('segmentation_data.txt', header=None)
classData = pd.read_csv('segmentation_classes.txt', sep="\t", header=None)
featureData = pd.read_csv('segmentation_names.txt', header=None)
DT = data
DT
```

```python
terms = featureData.ix[:,0]
terms
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
DT = min_max_scaler.fit_transform(DT)
DT
```

## B. Perform Kmeans

```python
kmeans = KMeans(n_clusters=7, max_iter=500, verbose=1) # initialization
kmeans.fit(DT)
```

```python
clusters = kmeans.predict(DT)
```

```python
print clusters
```

```python
print kmeans.cluster_centers_
```

```python
print "\t\t\t\tC0\tC1\tC2\tC3\tC4\tC5\tC6"
for i in range(len(terms)):
    print "%20s\t\t%.4f\t%.4f\t%.4f\t%.4f\t%.4f\t%.4f\t%.4f" %(terms[i],kmeans.cluster_centers_[0][i],kmeans.cluster_centers_[1][i],kmeans.cluster_centers_[2][i],kmeans.cluster_centers_[3][i],kmeans.cluster_centers_[4][i],kmeans.cluster_centers_[5][i],kmeans.cluster_centers_[6][i])
```

```python
t = list(classData.ix[:,1])
c = clusters
print completeness_score(t,c)
```

```python
print homogeneity_score(t,c)
```

## C. Perform PCA

```python
from sklearn import decomposition
```

```python
pca = decomposition.PCA(n_components=7)
DTtrans = pca.fit(DT).transform(DT)
```

```python
np.set_printoptions(precision=2,suppress=True)
print DTtrans
```

```python
print(pca.explained_variance_ratio_)
```

## D. Perform KMeans Again

```python
kmeans = KMeans(n_clusters=7, max_iter=500, verbose=1) # initialization
kmeans.fit(DTtrans)
```

```python
clusters = kmeans.predict(DTtrans)
```

```python
t = list(classData.ix[:,1])
c = clusters
print completeness_score(t,c)
```

```python
print homogeneity_score(t,c)
```
