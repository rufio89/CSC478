```python
import pylab as pl
import pandas as pd
import numpy as np
import apriori2
import apriori2 as ap
```

```python
playlist_data = pd.read_csv('playlists.txt', header=None)
song_data = pd.read_csv('song_names.txt',sep="\t", header=None)
```

```python
playlist_data
```

## A. Load Playlist Into Nested List

```python
new_list = []
for index, row in playlist_data.iterrows():
    for song in row:
        lst = song.split(' ')
        lst =  filter(None, lst)
        song_list.append(lst)
    new_list.append(lst)
    lst = []

playlist_nums = []
    
for num_list in new_list:
    t = [int(num) for num in num_list]
    playlist_nums.append(t)

playlist_nums
```

## B. Run Apriori on Playlist at Min-Support of .002

```python
L, support = ap.apriori(playlist_nums, 0.002)
```

## C. Generate Rules from Itemsets

```python
ruleList1 = ap.generateRules(L,support,metric='confidence', minMetric = 0.5)
```

```python
songs = list(song_data[1].ix[0:])
songs
song_hash = {}
for i in range(len(songs)):
    song_hash[i] = songs[i]

song_hash[1000]
```

```python
ap.pntRules(ruleList1, song_hash)
```

```python
ruleList2 = ap.generateRules(L,support,metric='lift', minMetric = 30.0)
```

```python
ap.pntRules(ruleList2, song_hash)
```
